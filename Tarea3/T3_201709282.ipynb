## Importar las librerias necesarias


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

# 1. Cargar los datos

```python
df = pd.read_csv('winequality-red.csv', sep=";")  # Cambia la ruta al CSV
```

# 2. Mostrar las primeras filas para inspección

```python
print(df.head())
print(df.dtypes)
```

# 3. Revisar información general

```python
df.info()
```

# 4. Resumen estadístico

```python
describe_df = df.describe()
print(describe_df)
```

# 5. Matriz de correlación

```python
plt.figure(figsize=(10, 6))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```

# 6. Gráficos
## Histograma de la calidad del vino

```python
sns.histplot(df['quality'], bins=6, kde=True)
plt.title('Distribución de la calidad del vino')
plt.show()
```

## Boxplot de alcohol vs calidad

```python
sns.boxplot(x=df['quality'], y=df['alcohol'])
plt.title('Distribución de alcohol según calidad')
plt.show()
```

## Relación entre ácido cítrico y calidad

```python
sns.scatterplot(x=df['citric acid'], y=df['quality'])
plt.title('Ácido cítrico vs Calidad')
plt.show()
```

# 7. Conclusiones

```python
conclusiones = """
1. La mayoría de los vinos tienen una calidad entre 5 y 6.
2. Existe una relación positiva entre el contenido de alcohol y la calidad del vino.
3. El ácido cítrico parece no tener una fuerte correlación con la calidad del vino.
"""
print(conclusiones)
```
